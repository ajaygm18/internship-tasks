        root1 = (-b + math.sqrt(det)) / (2 * a)

        print(f"one real root is {-b / (2*a)}")

        img = math.sqrt(abs(det)) / (2*a)
        print(f"Root1 {real} + {img}i")

```python
def check_prime(num: int):

    flag = 0

    for i in range(2, num //2):
        if num % i == 0 and i != num:
            flag = 1
            break

    if flag == 0:
        print("Prime")
    else:
        print("Not Prime")
        
check_prime(7)
        
```

```python
def chk_armst(num: int):
    orig = num
    total = 0
    lst = []
    while num != 0:
        digit = num % 10
        num = num // 10
        lst.append(digit)
        

    cnt = len(lst)

    for num in lst:
        total += num ** cnt
    if total == orig:
        print("Armstrong")
    else:
        print("Not Armstrong")


chk_armst(153)
```

```python
def check_neil_range():
    start, end = 1, 200
    for i in range(start, end):

        cnt = len(str(i))

        num = i
        total = 0
        while num != 0:
            digit = num % 10
            total += digit ** cnt
            num //= 10

        if i == total:
            print(i)

        
check_neil_range()
        

    
```

```python
def find_lcm():
    a, b = 54, 24
    if a > b:
        greater = a
    else:
        greater = b
    num = 0
    while True:
        num += greater
        if num % a == 0 and num % b == 0:
             print("LCM is", num)
             break
        
find_lcm()
```

```python
def utest():
    import unittest
    
    def add(a, b):
        return a + b
    
    def divide(a, b):
    
        if b == 0:
            raise ValueError("Can't Divide by Zero")
        return a / b
    
    class TestMathOperations(unittest.TestCase):
    
        def setUp(self):
            self.a = 5
            self.b = 10
            print("Setting UP resources")
    
        def tearDown(self):
            print("Cleaning up resources")
            del self.a, self.b
    
        def test_add(self):
            result = add(self.a, self.b)
            self.assertEqual(result, 15)
    
        def test_divide(self):
            result = divide(self.a, self.b)
            self.assertEqual(result, 0.5)
    
        def test_divide_by_zero(self):
            with self.assertRaises(ValueError):
                divide(5, 0)
    
    unittest.main()
    
```
