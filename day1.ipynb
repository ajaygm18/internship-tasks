```python
import math
def solve_quadratic():
    a = float(input("Coefficient a :"))
    b = float(input("Coefficient b :"))
    c = float(input("Coefficient c :"))

    det = (b ** 2) - (4 * a * c)

    if det > 0:
        root1 = (-b + math.sqrt(det)) / (2 * a)
        root2 = (-b - math.sqrt(det)) / (2 * a)
        print(f"root1 = {root1}")
        print(f"root2 = {root2}")

    elif det == 0:
        print(f"one real root is {-b / (2*a)}")
    else:
        real = -b / (2*a)
        img = math.sqrt(abs(det)) / (2*a)
        print(f"Root1 {real} + {img}i")
        print(f"Root2 {real} - {img}i")

solve_quadratic()
```

```python
def check_prime(num: int):

    flag = 0

    for i in range(2, num //2):
        if num % i == 0 and i != num:
            flag = 1
            break

    if flag == 0:
        print("Prime")
    else:
        print("Not Prime")
        
check_prime(7)
        
```

```python
def range_prime():
    a, b = 2, 20
    for i in range(a, b+1):
        flag = 0
        for j in range(2, i):
            if i % j == 0:
                flag = 1
                break
        if flag == 0:
            print(i)

range_prime()
```

```python
def chk_armst(num: int):
    orig = num
    total = 0
    lst = []
    while num != 0:
        digit = num % 10
        num = num // 10
        lst.append(digit)
        

    cnt = len(lst)

    for num in lst:
        total += num ** cnt
    if total == orig:
        print("Armstrong")
    else:
        print("Not Armstrong")


chk_armst(153)
```

```python
def check_neil_range():
    start, end = 1, 200
    for i in range(start, end):

        cnt = len(str(i))

        num = i
        total = 0
        while num != 0:
            digit = num % 10
            total += digit ** cnt
            num //= 10

        if i == total:
            print(i)

        
check_neil_range()
        

    
```

```python
def find_lcm():
    a, b = 54, 24
    if a > b:
        greater = a
    else:
        greater = b
    num = 0
    while True:
        num += greater
        if num % a == 0 and num % b == 0:
             print("LCM is", num)
             break
        
find_lcm()
```

```python
def find_hcf():

    a, b = 54, 24

    if a > b:
        smaller = b
    else:
        smaller = a

    for i in range(1, smaller):
        if a % i == 0 and b % i == 0:
            hcf = i
    print(hcf)

find_hcf()
```

```python
def calculator():

    def add(a, b):
        return a + b

    def sub(a, b):
        return a - b

    def mul(a, b):
        return a * b

    def div(a, b):
        return a / b

    def get_inp():
        a = float(input("Enter number 1: "))
        b = float(input("Enter number 2: "))
        return a, b

    while True:

        print("*"*20)
        print("1.Addition\n2.Subtraction\n3.Multiplication\n4.Division")
        inp = int(input("Enter your choice: "))
        
        
        match inp:
            case 1:
                x, y = get_inp()
                print(add(x, y))
                break
            case 2:
                x, y = get_inp()
                print(sub(x, y))
                break
            case 3:
                x, y = get_inp()
                print(mul(x, y))
                break
            case 4:
                x, y = get_inp()
                print(div(x, y))
                break
            case _:
                return
                

calculator()
    
```

```python
def fib_recursion(x):
    if x <= 1:
        return x
    
    else:
        return fib_recursion(x-1) + fib_recursion(x-2)
number_of_fib = 10
for i in range(number_of_fib):
    print(fib_recursion(i))
    
```

```python
def find_fact(x):
    if x <= 1:
        return x
    return x * find_fact(x - 1)
    
print(find_fact(5))
    
```

```python
def bmi(w, h):
    return round((w/(h**2)), 2)

def bmi_chk():
    weight = float(input("Enter the weight"))
    height = float(input("Enter the height"))

    x = bmi(weight, height)
    print(x)
    if x < 18.5:
        print("underweight")
    elif x < 24.9:
        print("normal")
    elif x < 29.29:
        print("Overweight")
    else:
        print("Obese")

bmi_chk()
```
