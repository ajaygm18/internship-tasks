```python
#decorator practice
import time

def time_wrapper(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("Function {} took {}sec's to execute".format(func.__name__, round(end - start, 4)))
    return wrapper
        
@time_wrapper
def test_sleeper():
    time.sleep(1)
    return "Done"

test_sleeper()
```

```python
def binary_search(lst, search):
    low, high = 0, len(lst) - 1
    while low <= high:
        mid = (low + high) // 2
        if lst[mid] == search:
            return mid
        elif lst[mid] > search:
            high = mid -1
        else:
            low = mid + 1

print(binary_search([1, 2, 5, 6, 9, 10, 12, 15, 18], 18) + 1)
```

```python
#instance method class method static method
class Car:
    wheels = 4
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    #instance method
    def drive(self):
        print("Driving smooth")

    
    @classmethod
    def drift(cls):
        print("Drifting")
        #class attributes can be accessed via cls keyword
        print(cls.wheels)

    @staticmethod
    def reverse():
        print("Reverse Gear")
        
    

car1 = Car("Toyota", 'x12')
car2 = Car("Hyunda1", 'xdx')

#instance method
car1.drive()

#class method
car1.drift()

#static method
Car.reverse()
```

```python
#Getters and Setters

class BankAccount:
    def __init__(self, name, bal=0):
        self.balance = bal
        self.name = name

    #getter
    def fetch_balance(self) -> int:
        return self.balance

    #setter
    def deposit(self, amount: int):
        if amount <= 0:
            return "deposit +ve numbers only"

        else:
            self.balance += amount
            return "Deposit Successful"


    def withdraw(self, amount: int):
        if amount <= 0:
            return "withdraw +ve numbers only"
            
        elif amount >= self.balance:
            return "Insufficient Balance"

        else:
            self.balance -= amount
            return "withdraw Success"

acc1 = BankAccount("Ajay", 30000)
acc2 = BankAccount("Amarnath")

print(acc1.fetch_balance())
print(acc2.fetch_balance())

print(acc1.withdraw(31))
print(acc2.deposit(70000))

print(acc1.fetch_balance())
print(acc2.fetch_balance())

print(BankAccount.mro())
BankAccount.__mro__
```

```python
#Polymorphism same methods behave differently when called through different classes

class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.l = length
        self.b = breadth
    
    def area(self):
        return self.l * self.b

    def perimeter(self):
        return 2 * (self.l + self.b)


class Circle(Shape):
    def __init__(self, radius):
        self.r = radius

    def area(self):
        return 3.14 * (self.r**2)

    def perimeter(self):
        return 2 * 3.14 * self.r


shapes = [Rectangle(5, 8), Circle(9), Rectangle(34, 71)]

for s in shapes:
    print(f"Area of {s.__class__.__name__} is {s.area()} and Perimeter is {s.perimeter()}")
        
```

```python
def digit_palindrome(num):
    original = num
    rev = 0
    while num != 0:
        digit = num % 10
        rev = (rev * 10) + digit
        num = num // 10
    return rev == original

digit_palindrome(212212)
    
```

```python
# sum of digits of a number

def sumofdigits():
    x = 34543
    sum = 0
    while x != 0:
        rem = x % 10
        x = x // 10
        sum += rem
    print(sum)

sumofdigits()

```

```python
def bin_to_decimal():
    x = 101
    sum = 0
    for i in range(len(str(x))):
        d = x % 10
        sum += 2 ** i * d
        x //= 10
    print(sum)

bin_to_decimal()
```

```python
def oct_to_decimal():
    x = 62
    sum = 0
    i = 0
    while x != 0:
        d = x % 10
        sum += (8 ** i) * d
        x //= 10
        i += 1
    print(sum)


oct_to_decimal()
```

```python
def hex_to_decimal():
    x = "1A5"
    i = len(x) - 1
    sum = 0
    j = 0
    while i >= 0:
        print(i)
        if (ord(x[i]) >= ord('0')) and (ord(x[i]) <= ord('9')):
           sum += (16 ** j) * int(x[i])
        elif (ord(x[i]) >= ord('A')) and (ord(x[i]) <= ord('F')):
            sum += (16 ** j) * (ord(x[i]) - 55)
        i-=1
        j+=1
    print(sum)
    
hex_to_decimal()
```
